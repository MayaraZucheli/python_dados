```python
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

selic_inicio = noticia.find('2,75%')
selic = noticia[selic_inicio:selic_inicio + 5]

ano_inicio = noticia.find('6')
ano = noticia[ano_inicio:ano_inicio + 1]

print(f"Taxa Selic: {selic}")
print(f"Ano: {ano}")
```
